## Edit data

ในที่นี้จะแสดงตัวอย่างการ edit ข้อมูลโดยใช้ฟังก์ชันที่มีใน pandas

#### ตัวอย่างการปรับเปลี่ยนข้อมูลสภาพครัวเรือน

ในส่วนนี้จะแสดงตัวอย่างการปรับเปลี่ยนข้อมูล โดยการแยกข้อมูลที่อยู่ในวงเล็บ ของคอลัมน์ *การใช้น้ำอุปโภค* ออกมาอยู่ในคอลัมน์ใหม่


##### แก้ไขชื่อคอลัมน์

```
fromGDrive.rename(columns={"จำนวนที่ดินทำกิน (ไร่)-เป็นเจ้าของ": "own_rai","จำนวนที่ดินทำกิน (ไร่)-เช่า": "rent_rai" },inplace = True)
# เปลี่ยนชื่อ Column ให้เป็น own-rai & rent-rai (columns เป็น dict มี key เป็นชื่อเดิม value เป็นชื่อที่ต้องการเปลี่ยน)
```

ใช้คำสั่ง regex ง่ายมาก แต่ต้องเรียนรู้ regex
คำสั่ง regex ให้ใส่ตัวอักษร r ก่อน จากนั้นใส่คำสั่งลงไประหว่าง '...' โดยไม่ต้องมี / ปิดหน้าปิดหลังเหมือน regex ปกติ

เรียนรู้เพิ่มเติมได้ที่ https://www.youtube.com/watch?v=rhzKDrUiJVk

```
fromGDrive['ความถี่การใช้น้ำ'] = fromGDrive['ข้อมูลการใช้น้ำอุปโภค'].str.extract(r'\(([^\)]+)\)')
# สร้าง column ใหม่ ชื่อ "ความถี่การใช้น้ำ" โดย extract ข้อมูลจากอีก Column ที่ชื่อ "ข้อมูลการใช้น้ำอุปโภค" แบบใช้ Regex
```

```
fromGDrive
```

##### ใช้คำสั่ง split เพื่อแยกตัวหนังสือออกจากกัน

ข้อมูลการใช้น้ำอุปโภค จะมีข้อมูลในวงเล็บอยู่ เราจะทำการแยกออกมา

```
fromGDrive[fromGDrive['ข้อมูลการใช้น้ำอุปโภค'].str.contains(')',na=False,regex=False)]
# เอา row ที่ อยู่ในเงื่อนไขที่ว่า column "ข้อมูลการใช้น้ำอุปโภค" มี วงเล็บ -> เพื่อที่เราจะทำการเเยกออกไปอีก Column
```

```
# ดูข้างในวงเล็บว่ามีอะไร -> พิจารณาเฉพาะ Column ที่มีวงเล็บข้างใน
fromGDrive['ข้อมูลการใช้น้ำอุปโภค'].str.contains(')',na=False,regex=False)
```

หลังจากนั้นใช้คำสั่ง split เพื่อแยกข้อมูลในวงเล็บออกไปอีกคอลัมน์

```
fromGDrive[['ข้อมูลการใช้น้ำอุปโภค','หมายเหตุ']] = fromGDrive['ข้อมูลการใช้น้ำอุปโภค'].str.split('(',expand=True)
# split column then add new column as "หมายเหตุ"
```

หลังแยกข้อมูลในวงเล็บออกไปจะได้ข้อมูลแบบด้านล่างนี้

```
fromGDrive.head()
```

##### การใช้คำสั่ง replace เพื่อลบตัวอักษรที่ไม่ต้องการ

ข้อมูลในคอลัมน์ใหม่ที่ชื่อ *หมายเหตุ* จะยังมีวงเล็บปิดติดอยู่ ในส่วนนี้จะทำการลบวงเล็บออกโดยใช้คำสั่ง replace

```
fromGDrive['หมายเหตุ']=fromGDrive['หมายเหตุ'].str.replace(')','',regex=False) # delete ) by using regex
```

## Transform pivot(wide) table or long table

แนะนำให้ดู [Pivot and Unpivot Dataframe](https://www.youtube.com/watch?v=abO6e_b-EHs)

Library ในการแสดงผลหรือการวิเคราะห์จะต้องการรูปแบบตารางที่ต่างกัน บางอย่างต้องการข้อมูลที่เรียงตัวแบบ long หรือ 1 คอลัมน์ต่อ 1 ตัวแปร บางอย่างต้องการข้อมูลที่เรียงตัวแบบ wide หรือ pivot หรือตัวแปรที่เป็นตัวแปรเชิงกลุ่มกลายเป็นแต่ละคอลัมน์

ในตัวอย่างต่อไปนี้ จะทดลองแปลงค่าฝุ่นที่อยู่ในตาราง fromGithub ไปเป็นตารางแบบ pivot

##### คำสั่ง pivot

อยากรู้ว่าแต่ละวันมีค่าเหมือนกันหรือไม่ โดยเอาชั่วโมงเดียวกันมาเทียบกัน ต้องทำ pivot นั่นก็คือให้แต่ละคอลัมน์เป็นชั่วโมง และแต่ละแถวเป็นเวลา

ในการทำ pivot นั้น เราจะต้องทำการกำหนดว่าอะไรจะเป็นแถว หรือ index และอะไรเป็นคอลัมน์ หรือ columns โดย index จะต้องเป็น unique หรือต้องไม่มีการซ้ำกัน

```
# Pivot 1 : สร้าง DataFrame แยกตามรายชั่วโมง
pivot_1 = s102.pivot(index='Time',columns='h',values='Value') # index เป็น วัน (row) , column เป็น h , แทน value ลงไปในเเต่ละชอง
```

```
pivot_1
```

อีกตัวอย่างคือทำการหาค่าเฉลี่ยรายวันของแต่ละเซ็นเซอร์ก่อน จากนั้นทำ pivot table เพื่อเปรียบเทียบค่าเฉลี่ยรายวันของแต่ละเซ็นเซอร์

```
dailyavg = fromGithub.groupby(['Series','Time','SensorGroup'])['Value'].mean().reset_index()
# หา mean ของเเต่ละ row โดย groupby Series , Time (คำนวณอิงตามวัน) (เเสดงใน DataFrame)
```

```
dailyavg
```

```
# Pivot 2 : DataFrame อิงตามชื่อ Sensor โดยเเยกตาม Time (plot มาจาก dailyavg)
pivot_2 = dailyavg.pivot(index='Time',columns='Series',values='Value')
```

```
pivot_2
```

ถ้าเราต้องการจะแปลงกลับไปเป็น long table ให้ใช้คำสั่ง stack

##ฟังก์ชัน apply, assign และ lambda

ถ้าจะทำอะไรที่แปลก ๆ หน่อย อย่าเพิ่งวน loop ให้ลองใช้ฟังก์ชั่นเหล่านี้ในการจัดการ

ฟังก์ชัน apply เป็นฟังก์ชันที่ใช้สำหรับรันฟังก์ชั่นอื่น ๆ โดยจะกำหนดได้ว่าจะรันฟังก์ชั่นนั้น ๆ ในแกนคอลัมน์ (axis=1) หรือรันในแกนแถว (axis=0)

ฟังก์ชัน assign จะเป็นฟังก์ชันที่ใช้สำหรับรันฟังก์ชันอื่น ๆ เช่นกัน แต่จะนำผลไปใส่ในคอลัมน์ใหม่หรือแก้ไขคอลัมน์ที่มีอยู่แล้ว

ทั้งฟังก์ชัน apply และ assign จะใช้ร่วมกับฟังก์ชันอะไรก็ได้ หรือใช้กับฟังก์ชัน lambda ซึ่งก็คือฟังก์ชันที่เรากำหนดขึ้นมาเอง

lambda เป็นฟังก์ชันที่ไม่เหมือนฟังก์ชันดั้งเดิมธรรมดา ก็คือไม่ต้องตั้งชื่อ แต่มีขั้นตอนซับซ้อนไม่ได้
