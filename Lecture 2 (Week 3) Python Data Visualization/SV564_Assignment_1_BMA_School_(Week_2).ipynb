# Task 3 : แสดงสัดส่วนนักเรียนต่อครู (แยกตามอะไรก็ได้ ให้ทราบแนวโน้มที่น่าสนใจ)

```python
# ผมมี Idea ว่าอยากทำเป็น Box Plot เพราะจะได้เห็นว่าข้อมูลไหนเป็น Outlier (ครูขาดเเคลน หรือ ครูมากเกิน)

# Step 1 : หา DataFrame ที่เเสดง จน. student และ teacher แบ่งตามชื่อโรงเรียนในกทม. (เลือก Column ที่จำเป็นออกมาก่อน)
  # เพื่อที่จะดูว่า มีกี่โรงเรียนขาดเเคลนบุคลากร , บุคลากรครูล้น

import matplotlib.pyplot as plt

stu_teach_data = data_bma.loc[ : , ["name" , "dname" , "num_stu" , "teacher"] ]
  # เลือก Column ที่ชื่อ dname (เขต) , num_stu , teacher ออกมา

stu_teach_data

# Step 2 : สร้าง Column Ratio_student_to_teacher ขึ้นมา (อจ. 1 คน ต่อ นักเรียน x คน)
## Syntax : df[...] = df[..] / df[...] เสมือนเป็นการสร้าง column ใมห่
stu_teach_data["Ratio_Student_to_Teacher"] = stu_teach_data["num_stu"] / stu_teach_data["teacher"]\

# ลองเรียก dataFrame ใหม่ดู
stu_teach_data
```

```python
# Step 3 : ผมจะเอาแต่ละโรงเรียนมา Plot เป็น Box Plot เพื่อดูว่ามีกี่โรงเรียนที่เกิน Outlier
import seaborn as sb
import matplotlib.pyplot as plt

# 3.1) Import Seaborn เพื่อ plot box plot
plt.figure(figsize=(8,6)) # กำหนดขนาด พื้นหลัง Graph

# 3.2) Plot Box Plot
sb.boxplot(data = stu_teach_data, y = "Ratio_Student_to_Teacher") # Plot ค่าจาก DataFrame โดยแกน y ให้เป็น Ratio นร. ต่อ อจ.
plt.title("Graph 3 : Box Plot แสดงจำนวน Ratio นักเรียนต่ออาจารย์ ในแต่ละโรงเรียน")

# 3.3) Plot Strip Plot (Plot จุดข้อมูล เพื่อดูว่าแนวโน้มการกระจุกตัวอยู่ที่ไหนเป็นหลัก)
sb.stripplot(data = stu_teach_data , y = "Ratio_Student_to_Teacher" , color = "green" , size = 5) # ให้เป็นสีเขียว จนาด 5px
plt.show()
```

ทีนี้ผมจะลองเอาความรู้ Stat ตอน ม.ปลาย มาลองคำนวณเพื่อหาว่า โรงเรียนไหนบุคลากรครู มากเกินไป หรืออยู่ในช่วงที่ "ขาดแคลน"

```python
# คำนวณจาก Data Frame เพื่อหา Quartile 1 และ 3 ออกมา
Q1 = stu_teach_data["Ratio_Student_to_Teacher"].quantile(0.25) # ค่าที่แสดงว่าเป็น 25% แรกของข้อมูล
Q3 = stu_teach_data["Ratio_Student_to_Teacher"].quantile(0.75) # ค่าที่แสดงว่าเป็น 25% ท้ายของข้อมูล

# ผมจะลองคำนวณ IQR (ค่าส่วนต่าง Q3 กับ Q1 พูดง่ายๆคือ Bound ที่รับได้ของข้อมูล) แล้วจะลองคำนวณหา Lower Bound กับ Upper Bound เพื่อกั้นให้ค่าที่ดีหรือเเย่กว่านี้เป็น Outlier
# โดยๆๆๆ Lower = Q1 - 1.5 IQR , Upper = Q3 + 1.5 IQR
IQR = Q3 - Q1
lowerBound = Q1 - (1.5 * IQR)
upperBound = Q3 + (1.5 * IQR)

# หา Outlier โดยใส่ Condition เข้าไปในการเลือกข้อมูลใน DataFrame
outlier_ratio = stu_teach_data[
          (stu_teach_data['Ratio_Student_to_Teacher'] < lowerBound) | (stu_teach_data["Ratio_Student_to_Teacher"] > upperBound)
          ]
  # เลือก Row ที่่ Ratio < Lower Bound และ Ratio > Upper Bound

# ตอนนี้ผมได้ Data ของโรงเรียนที่เป็น Outlier แล้ว ก็จะลองแสดงชื่อโรงเรียน และ จน.โรงเรียนออกมา
print("จำนวนโรงเรียนขาดแคลนครู (Outliers):", len(outlier_ratio))
outlier_ratio[ ["name", "dname", "Ratio_Student_to_Teacher"] ]
```
