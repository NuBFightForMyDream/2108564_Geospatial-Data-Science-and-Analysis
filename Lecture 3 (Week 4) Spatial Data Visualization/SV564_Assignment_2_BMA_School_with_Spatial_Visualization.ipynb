```python
# Step 1 : Import BMA_data and call DataFrame
import pandas as pd
import folium as fofo

BMA_school_spatial = pd.read_csv("/content/drive/MyDrive/CU Intania Starter Kit Chatrphol/Year 2 1 Electrical Engineering ⚡/2108564 Geospatial Data Science and Analysis/SV564 Lecture 3 Spatial Data Visualization with Folium and GeoPandas/SV564_Lecture3_BMASchoolSpatial.csv")
BMA_school_spatial # call DataFrame
```

```python
# Step 2 : create Map object & create needed elements

## Part 1 : create list
latitudes = list(BMA_school_spatial["lat"]) # list of latitudes (create from dataFrame column)
longitudes = list(BMA_school_spatial["lng"]) # list of latitudes (create from dataFrame column)
start_location = [ latitudes[0] , longitudes[0] ]

## Part 2 : create Map
mapBMA = fofo.Map( # attribute inside : location (ตั้งเป็น start_location), zoom_start (ตั้งเป็น 10)
          location = start_location ,
          zoom_start = 15)

## Part 3 : add marker to each school
for (each_lat , each_lng) in zip(latitudes , longitudes) :

  # put marker using fofo.marker( attribute ).add_to(map)
  fofo.Marker(
      location = [each_lat , each_lng] ,
      icon = fofo.Icon(color = "red" , icon = "home")

      ).add_to(mapBMA)

## call mapBMA if data updated (icon ของเเต่ละ location ถูกใส่ลงไปลงไปใน map เเล้ว)
mapBMA
```

```python
# Step 3 : Put Datas into popup in Map

## Part 1 : create some necessary elements to put in Pop-Up in Map
latitudes = list(BMA_school_spatial["lat"]) # list of latitudes (create from dataFrame column)
longitudes = list(BMA_school_spatial["lng"]) # list of latitudes (create from dataFrame column)

school_names = list(BMA_school_spatial["name"])
dnames = list(BMA_school_spatial["dname"])
num_stu_allschool = list(BMA_school_spatial["num_stu"])
teacher_allschool = list(BMA_school_spatial["teacher"])

# Part 2 : For loop , add each element to pupup
for pos in range(len(school_names)):

    # ลองใส่ข้อความในรูปแบบ HTML เพราะ Folium based on HTML ถ้าใส่เป็น \n จะไม่ขึ้น new line ให้ เลยใส่ <br> เสมือน new line
    popup_text = (
        f"ชื่อโรงเรียน: {school_names[pos]}" + '<br>' +
        f"เขต: {dnames[pos]}" + '<br>' +
        f"จำนวนนักเรียน: {num_stu_allschool[pos]}" + '<br>' +
        f"จำนวนครู: {teacher_allschool[pos]}"
    )

    # ใส่ marker ทับอันเดิมไปเลย (ลองใส่เเบบ marker เดิม เเล้ว update เเต่ไม่ขึ้น pop-up TwT)
    marker = fofo.Marker(
        location = [latitudes[pos], longitudes[pos]], # update ตำเเหน่ง
        popup = fofo.Popup(popup_text , max_width = 500) ,
        icon = fofo.Icon(color="red", icon="home")
    )
    marker.add_to(mapBMA) # add กลับไปที่ map
```

```python
# ลองเช็คดูว่า map มีการ update ยัง (update แล้ว เย่ๆๆๆ)
mapBMA
```
